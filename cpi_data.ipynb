```python
import pandas as pd
import utils
```

```python
engine = utils.get_engine()
```

```python
data = pd.read_html("https://www.bls.gov/charts/consumer-price-index/consumer-price-index-by-category-line-chart.htm")
data = data[0]
```

```python
data.head()
```

```python
data.columns = [c.lower().replace(' ', '_').replace('-', '_') for c in data.columns]
```

```python
data['date'] = pd.to_datetime(data.month)
```

```python
data.columns
```

```python
selected_columns = list(set(data.columns)-set(['date', 'month']))
```

```python
data[selected_columns] = data[selected_columns].apply(lambda x: x.str.rstrip("%").astype(float)/100, axis=1)
```

```python
data.set_index('date', inplace=True)
```

```python
data[selected_columns].to_sql('cpi_data',
                              schema='public',
                              index_label='date',
                              if_exists='append',
                              con=engine)
```
